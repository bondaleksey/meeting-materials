# Тема: разбор кейсов хакатона

# 1. Список кейсов

## 1.1 Case 1 Генерация тегов для видео

На платформу Rutube ежедневно загружаются сотни тысяч видео, большая часть которых — это контент, созданный пользователями (UGC). Помимо этого, присутствует популярный лицензионный контент, такой как шоу и передачи. Для упорядочивания этого обширного контента необходимо разработать систему тегирования, которая разделяет видео по категориям и подкатегориям. Такая система облегчит навигацию и улучшит рекомендательную систему, позволяя предлагать пользователям контент из их любимых категорий.

Участникам необходимо создать прототип системы, который будет генерировать теги для видеоконтента на основе видео, его названия и описания. Тегирование должно происходить по универсальному иерархическому списку тегов, предоставляемому кейсодержателем и охватывающему различные тематики и подтематики. На выходе система должна предоставлять список вероятных категорий и подкатегорий, к которым относится видео.

## 1.2 Case 2 Генерация виральных клипов

Сотрудники, занимающиеся продвижением оригинального контента, часто вручную извлекают ключевые моменты из длинных видео для создания коротких клипов, которые должны заинтересовать зрителя и побудить его к просмотру полного видео. Этот процесс трудоемок и требует значительных временных затрат. Необходимо автоматизировать эту задачу с помощью ИИ, чтобы помочь редакторам контента эффективно генерировать виральные короткие видео.

Участникам хакатона предстоит разработать сервис, который позволит загружать длинные видео и автоматически генерировать из них набор из коротких вертикальных клипов с ключевыми моментами. Эти клипы должны быть способны максимально завлечь зрителя и привести к просмотру оригинального контента.

## 1.3 Case 3 Интеллектуальный помощник оператора службы поддержки

Видеохостинг RUTUBE ежедневно сталкивается с тысячами пользовательских обращений, связанных с различными аспектами использования сервиса. В настоящее время каждое из таких обращений обрабатывается вручную операторами службы поддержки, что приводит к значительным временным и денежным затратам.

Участникам предстоит разработать систему интеллектуального помощника для генерации ответов на основе базы знаний часто задаваемых вопросов и целевых ответов на них. Система должна демонстрировать высокую точность и скорость работы, а также быть интегрирована с Telegram ботом для улучшения пользовательского взаимодействия.

## 1.4 Case 4 Предсказание социально-демографических характеристик пользователя

Пользователи RUTUBE не всегда указывают свои данные, такие как возраст и пол, что затрудняет формирование портрета пользователя и создание персонализированных рекомендаций. Это ограничивает возможности платформы в предоставлении контента, который наиболее подходит интересам и потребностям пользователей, тем самым ухудшая пользовательский опыт.

Необходимо разработать модель, которая на основе истории просмотров сможет предсказывать пол и возраст пользователя. Подобные данные смогут служить сильными признаками в моделях рекомендательной системы, поиска и модерации контента платформы.

## 1.5 Case 5 Решение проблемы холодного старта для новых пользователей

Новые пользователи видеохостинга RUTUBE сталкиваются с обширным и разнообразным контентом, среди которого точно есть то, что пользователю может быть интересно. Однако без какой-либо информации об исторических просмотрах сложно определить, что можно показать пользователю в рекомендациях для наиболее точного попадания в его интересы. Проблема "холодного старта" для новых пользователей является критически важной для обеспечения их положительного первого опыта и последующей активности на платформе.

В рамках кейсового задания участникам предстоит разработать алгоритм, способный после нескольких показов рекомендаций и считывания реакции на них, предсказывать интересы новых пользователей, несмотря на отсутствие их истории активности. Итоговый продукт должен быть представлен в виде веб-приложения, где пользователи могут давать обратную связь и получать обновленные рекомендации.

## 1.6 Case 6 Поиск дубликатов видео

На платформе коротких видео Yappy острой проблемой является наличие дубликатов контента, которые могут нарушать авторские права, снижать охваты оригинальных видео и ухудшать пользовательский опыт. Сложность проблемы заключается в большом объеме новых видео, скорости их обработки и разнообразии способов модификации контента, что делает ручное определение дублей практически невозможным.

Участникам предлагается создать микросервис, который с использованием технологий искусственного интеллекта будет определять дубликаты публикуемых видео. Ожидаемый результат — это прототип системы, включающей в себя алгоритмы анализа видео, базу данных для хранения обработанных роликов и API для проверки новых видео на наличие дублей.

Метрика будет. Лидерборд будет.
